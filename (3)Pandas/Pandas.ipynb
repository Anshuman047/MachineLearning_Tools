### Handelling NAN entries EDA
* Either we would enter some data in it or remove some rows, we generally donot leave them as such
* Drop NAN or fill NAN, operations to be carried out
* Try to drop or delete the row as data are very important

```python
print(df.head())
df.iloc[2:4,1:3]=np.nan # Made some entries in the row=0 or an constant provided by numpy that is nan means empty
df.head()
```

```python
df.describe()
```

```python
#Lets see how to handle it:
#1) Dropna()
df.dropna(inplace=True)# It will drop the whole-rows having the Nan,whenever use drop always pass arg inplace=True for changes having in the original 
df.head()
```

```python
#Hence now our labelled-index is miss-arranged so, drop them and add new arranged one
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
#Again make the NAN for checking the 2nd function
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
#Adding items at NAN in PetalLength coloumn
#Same procedure as above
df.PetalLength.fillna(df.PetalLength.mean(),inplace=True)
df.head()
```

```python
#Adding items at NAN in SepalLength coloumn
#Whatever mean of that column,fill those at NAN
df.SepalWidth.fillna(df.SepalWidth.mean(),inplace=True) #Will fill NAN in SepalWidth coloumn with its coloumn mean
#The above also wonot happen in inplace so made it true in it
df.head()
```

```python
#Again make the NAN for checking the below segement answer
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
#Or else we could insert the mean of same flower type in respective col, rather than taking all means
a=df[df.FlowerType=="Iris-setosa"]#This would segregrate those data of respective flower type in the dataFrame
print(a.tail())#Now a is also a new dataframe,having only 46 rows
#Now Mean would be find out from respective rows. which already have same flowertype
MeanSW=a.SepalWidth.mean()
MeanPL=a.PetalLength.mean()
print(df.head())
df.SepalWidth.fillna(MeanSW,inplace=True)
df.PetalLength.fillna(MeanPL,inplace=True)
df.head()
```

### Handling with strings based data:
* Its really hard to do any calculation with strings, generally we required numeric values

```python
df["Gender"]="Female"#Added a newColoumn named Gender and enter all Female strings in it 
df.head()
df.iloc[2:7,6]="Male"#Changed some Gender as Male
df.head()
```

```python
#Lets choose gender column and apply a function
#Function defined
def f(s):
    if(s=="Male"):
        return 1
    else:
        return 0
    
#Choosed the gender coloumn and applied the function stored the return values in another new coloumn named sex 
df['Sex']=df.Gender.apply(f)
df.head()

#Now gender column has no used as the string in it changed into repspective numerical
#Delete gender col:
#del df["Gender"]
```
